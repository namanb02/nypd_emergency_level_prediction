## <font color='crimson'>**Milestone - Feature Engineering**</font>

>> * Features involved => 'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED'
>
> * Finding Crash Hour
>
>> * Features involved => 'CRASH TIME'

---
Number of Vehicles involved

```python
def count_vehicles_involved(data: pd.DataFrame) -> pd.DataFrame:
    """
    Calculates the number of vehicles involved in each crash by counting
    the number of non-missing values in the specified vehicle code columns.
    """

    # 1. Define the relevant vehicle columns
    vehicle_cols = [
        'VEHICLE TYPE CODE 1',
        'VEHICLE TYPE CODE 2',
        'VEHICLE TYPE CODE 3',
        'VEHICLE TYPE CODE 4',
        'VEHICLE TYPE CODE 5'
    ]

    # 2. Select only the vehicle columns
    vehicle_data = data[vehicle_cols]

    # 3. Check for non-missing values (returns True/1 for non-NA, False/0 for NA)
    # The .notna() method is the robust way to check for anything but NA.
    is_involved = vehicle_data.notna()

    # 4. Sum the True (1) values across the rows (axis=1)
    # The sum of booleans (True=1, False=0) gives the total count.
    data['VEHICLES INVOLVED'] = is_involved.sum(axis=1)

    return data


data = count_vehicles_involved(data)
```

```python
conditions_emergency = [
    data['NUMBER OF PERSONS KILLED'] > 0,

    data['NUMBER OF PERSONS INJURED'] > 0
]

choices_emergency = [
    2,
    1
]

data['DEGREE OF EMERGENCY'] = np.select(
    conditions_emergency,
    choices_emergency,
    default=0
)
```
